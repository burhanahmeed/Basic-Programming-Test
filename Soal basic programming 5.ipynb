```python
def getFiction (n , m, arr) :
    i = 0
    newarr = []
    while i < n:
        k = 0
        new_alp = ''
        for k in arr[i]:
            if k.isalpha():
                new_alp += k
        if new_alp != '':
            newarr.append(new_alp)
        i = i + 1
    newarr = sorted(newarr)
    distinc = list(dict.fromkeys(newarr))
    contes = 0
    for i in distinc:
        if len(i) < 2:
            print(str(i) + ' ' + str(newarr.count(i)))
        else:
            contes = 1 + contes
    print('contested '+ str(contes))
```
