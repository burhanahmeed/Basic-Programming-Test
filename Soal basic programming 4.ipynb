There is a well-known puzzle called Word Search that involves looking for words in a grid of letters. The words are given in a list and can appear in the grid horizontally, vertically, or diagonally in any direction. In this task, you should implement a solver for word search. You will be given grids and a word to search for, and you have to find how many times that word comes out in the grid. Words that are spelled the same backwards and forwards, also known as palindromes, will not be given, so you don’t need to worry about words that match in the exact same spot in two different directions.

Input:

The first line is the number of test cases T. Each test case will have two numbers N and M, each on their own line given in that order. Following that is N lines of M lowercase letters each representing the grid of letters. Lastly, a word W is given that you must look for.

Output:

For each test case, output one line of the form “Case C: X” (without the quotes), where C is the case number (starting from 1), and X is how many times the word W appeared in the grid.

Constraints:

1 ≤ T ≤ 100
1 ≤ N ≤ 100
1 ≤ M ≤ 100
1 ≤ length(W) ≤ 100

Sample Output:

Case 1: 4
Case 2: 8

```python
def swap(string)->str:
    if len(string) == 0:
        return ''
    elif len(string) == 1:
        return string
    else:
#         print(swap(split(string)) + str(string[0]))
        return swap(split(string)) + str(string[0])
```

```python
def swap1(string)->str:
    if len(string) == 0:
        return ''
    elif len(string) == 1:
        return string
    else:
        return string[::-1]
```

```python
def split (string)->str:
    newstr = ''
    i = 0    
    while i < (len(string)-1):
        newstr = newstr + str(string[i+1])        
        i = i + 1
    return newstr
```

```python
def findHorizontal(word, searchWord: str, n: int, m: int) -> int:
    num = 0
    i = 0    
    while i < (n-1) :
        x = 0
        while x+len(searchWord) <= m:
            string = ''
            k = 0
            while k < len(searchWord):            
                string = string + word[i][x+k]
                k = k + 1
            if string == searchWord:
                num = num + 1
            if string == swap(searchWord):
                num = num + 1
            x = x + 1
        i = i+1
    return num
```

```python
def findVertical (word, searchWord: str, n, m)->int:
    num = 0
    i = 0
    while i < (m-1) :    
        x = 0
        while x+len(searchWord) <= n: 
            string = ''
            k = 0
            while k < len(searchWord):            
                string = string + word[x+k][i]
                k = k+1
            if string == searchWord:
                num = num + 1
            if string == swap(searchWord):
                num = num + 1
            x = x + 1
        i = i + 1
    return num
```

```python
def findWordDiagonal (word, searchWord: str, n, m)->int:
    num = 0
    u = 0
    while u < n:   
        i = 0
        while i < m :     
            if u+len(searchWord) <= n and i+len(searchWord) <= m:
                string = ''
                k = 0
                while k < len(searchWord):                
                    string = string + word[u+k][i+k]
                    k = k+1
#                 print(string)
                if string == searchWord:
                    num = num + 1
                if string == swap(searchWord):
                    num = num + 1
            i = i+1
        u = u+1
    return num
```

```python
def findDiagonalTop (word, searchWord: str, n, m)->int:
    num = 0
    u = 0
    while u < n:    
        i = 0
        while i < m:        
            if (u+1)-len(searchWord)  >= 0 and i+len(searchWord) <= m:
                string = ''
                k = 0
                while k < len(searchWord):                
                    string = string + word[u-k][i+k]
                    k = k+1
                
                if string == searchWord:
                    num = num + 1
                if string == swap(searchWord):
                    num = num + 1
            i = i+1
        u = u+1
    return num
```

```python
def convert(list):       
    # Converting integer list to string list 
    s = [str(i) for i in list]       
    # Join list items using join() 
    # res = int("".join(s)) 
    res = str("".join(s))
      
    return res
```

```python
def main (words, sw, n, m) :
#     return words[0][2]
#     print(sw)
#     print(n)
#     print(m)
#     print(words)
    num = findHorizontal(words, sw, n, m)
    num = num + findVertical(words, sw, n, m)
    num = num + findWordDiagonal(words, sw, n, m)
    num = num + findDiagonalTop(words, sw, n, m)
    return num    
```
