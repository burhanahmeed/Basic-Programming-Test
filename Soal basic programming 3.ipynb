```python
def pembagi1 (a, b, k) :
    num = range(a, b+1)
    angka = 0
    for i in num:
        if i % k == 0:
            angka = angka + 1
        return angka

# print("Case :"+ str(pembagi(8, 20, 4)))
```

```python
def pembagi (array) :
    a = int(array[0])
    b = int(array[1])
    num = range(a, b+1)
    angka = 0
    for i in num:
        if i % int(array[2]) == 0:
            angka = angka + 1
    return angka

print("Case :"+ str(pembagi(["1", "19", 4])))
```

```python
f = open('input.txt')
narr = []
for line in f:    
    narr.append(line)

theList = narr[1:]
N = 3

grouped = [theList[n:n+N] for n in range(0, len(theList), N)]

# grouped
for idx, i in enumerate(grouped):
    res = pembagi(i)
    print("Case "+ str(idx+1)+": " +str(res))
#     print("Case "+ str(idx+1)+": " + str(i)) 
grouped
```
